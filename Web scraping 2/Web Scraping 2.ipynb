```python
# Find the table containing the video details
table = driver.find_element(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"][1]')

# Find all rows in the table
rows = table.find_elements(By.TAG_NAME, 'tr')

# Scrape the details from each row
for row in rows[1:]:  # Exclude the header row
    cells = row.find_elements(By.TAG_NAME, 'td')
    # Check if the expected number of cells is present
    if len(cells) >= 5:
        rank = cells[0].text
        name = cells[1].text
        artist = cells[2].text
        upload_date = cells[4].text
        views = cells[3].text

    # Print the scraped details
    print(f"Rank: {rank}")
    print(f"Name: {name}")
    print(f"Artist: {artist}")
    print(f"Upload Date: {upload_date}")
    print(f"Views: {views}")
    print()

# Close the browser
driver.quit()
```

```python
# Load the web page
url = 'https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare'
driver.get(url)
```

```python

# Find the table rows containing the book deTails
table_rows = driver.find_elements(By.XPATH, '//table[@class="in-article sortable"]')
```

```python

# Initialize lists to store the scraped details
book_names = []
author_names = []
volumes_sold = []
publishers = []
```

```python
# Find the table rows
rows = driver.find_elements(By.XPATH, "//table[@class='in-article sortable']/tbody/tr")

# Extract data from each row
for row in rows:
    cells = row.find_elements(By.XPATH, "./td")
    book_names.append(cells[1].text)
    author_names.append(cells[2].text)
    volumes_sold.append(cells[3].text)
    publishers.append(cells[4].text)

# Create a DataFrame
data = {
    "Book Name": book_names,
    "Author Name": author_names,
    "Volumes Sold": volumes_sold,
    "Publisher": publishers
}
df = pd.DataFrame(data)

# Print the DataFrame
print(df)

# Close the browser
driver.quit()
```
