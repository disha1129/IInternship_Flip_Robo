```python
plt.figure(figsize=(10,10))

correlation_matrix= sns.diverging_palette(10, 220, as_cmap=True)

sns.heatmap(data=case.corr(), vmin=-1.0, vmax=1.0, center=0,  linecolor='k', annot=True, fmt='.3f',cmap=correlation_matrix)
plt.title('Correlarion Matrix')
plt.show()
```

```python
correlation_matrix = case.corr()

# Get the number of features
num_features = len(correlation_matrix.columns)

# Initialize lists to store positive and negative correlated features
positive_correlation_features = []
negative_correlation_features = []

# Loop through the correlation matrix
for i in range(num_features):
    for j in range(i + 1, num_features):
        corr_value = correlation_matrix.iloc[i, j]

        # Define a threshold for considering correlations
        threshold = 0.5  # You can adjust this threshold as needed

        # Check if the correlation is positive
        if corr_value >= threshold:
            positive_correlation_features.append((correlation_matrix.columns[i], correlation_matrix.columns[j], corr_value))
        # Check if the correlation is negative
        elif corr_value <= -threshold:
            negative_correlation_features.append((correlation_matrix.columns[i], correlation_matrix.columns[j], corr_value))

```

```python
# Print the lists of features with positive and negative correlations
print("Features with Positive Correlations:")
for feature_pair in positive_correlation_features:
    print(f"{feature_pair[0]} and {feature_pair[1]}: {feature_pair[2]}")

print("\nFeatures with Negative Correlations:")
for feature_pair in negative_correlation_features:
    print(f"{feature_pair[0]} and {feature_pair[1]}: {feature_pair[2]}")
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X_scale = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
# Creating a PCA instance
pca = PCA(random_state=1)

# Fitting PCA on the scaled data
pca.fit(X_scale)

# Extracting explained variance ratios
var_exp = pca.explained_variance_ratio_

# Calculating cumulative explained variance
cum_var_exp = np.cumsum(var_exp)

# Creating a DataFrame for explained variances of each component
var_ = pd.DataFrame(pca.explained_variance_.round(2),
                   index=["PC" + str(i) for i in range(1, len(var_exp) + 1)],
                   columns=["Explained Variance"])

# Printing the DataFrame
print(var_.T)

# Plotting explained variance ratios
plt.figure(figsize=(12, 7))
plt.bar(range(1, len(var_exp) + 1), var_exp,
        align='center', label='Individual Variance', color='teal', alpha=0.8)
plt.step(range(1, len(cum_var_exp) + 1), cum_var_exp,
         where='mid', label='Cumulative Variance', color='red')
plt.ylabel('Explained Variance Ratio')
plt.xlabel('Principal Components')
plt.xticks(np.arange(1, len(var_exp) + 1, 1))
plt.legend(loc='center right')
plt.show()

```

##### Conclusion
* The first 1 to 7 components account for about 99% of the total explained variance.
* We can run the analysis  with these 7 components.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

best_R2_score = 0
best_random_state = 0
num_iterations = 200

for random_state in range(1, num_iterations + 1):
    X_train, X_test, Y_train, Y_test = train_test_split(principle_x, Y, random_state=random_state, test_size=0.25)
    
    lin_reg = LinearRegression()
    lin_reg.fit(X_train, Y_train)
    
    y_pred = lin_reg.predict(X_test)
    
    R2 = r2_score(Y_test, y_pred)
    
    if R2 > best_R2_score:
        best_R2_score = R2
        best_random_state = random_state

print('Best R2 Score is', best_R2_score, 'on Random_state', best_random_state)
```

```python
import pandas as pd

# Create empty lists to store results
model_names = []
mae_list = []
mse_list = []
rmse_list = []
r2_list = []
cv_score_list = []

# Iterate through each model
for model in models:
    model.fit(X_train, Y_train)
    y_pred = model.predict(X_test)
    
    # Calculate error metrics
    mae = mean_absolute_error(Y_test, y_pred)
    mse = mean_squared_error(Y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(Y_test, y_pred)
    
    # Cross-Validation
    cv_scores = cross_val_score(model, principle_x, Y, cv=4)
    cv_score = cv_scores.mean()
    
    # Append results to lists
    model_names.append(str(model))
    mae_list.append(mae)
    mse_list.append(mse)
    rmse_list.append(rmse)
    r2_list.append(r2)
    cv_score_list.append(cv_score)

# Create a DataFrame
results_df = pd.DataFrame({
    'Model': model_names,
    'MAE': mae_list,
    'MSE': mse_list,
    'RMSE': rmse_list,
    'R2 Score': r2_list,
    'CV Score': cv_score_list
})


```
